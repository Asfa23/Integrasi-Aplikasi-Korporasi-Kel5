```python
import pandas as pd
```

# **Kapal Domestik**

```python
kapal_domestik = pd.read_excel('Jumlah Penumpang Domestik berdasarkan Moda Transportasi Kapal menurut provinsi, 2022-2023.xlsx', header=3)
kapal_domestik.columns = ['Provinsi', 'Datang_2022', 'Datang_2023', 'Berangkat_2022', 'Berangkat_2023']
kapal_domestik.head()
```

```python
kapal_domestik_datang = kapal_domestik[['Provinsi', 'Datang_2022', 'Datang_2023']]
kapal_domestik_datang.head()
```

```python
kapal_domestik_berangktat = kapal_domestik[['Provinsi', 'Berangkat_2022', 'Berangkat_2023']]
kapal_domestik_berangktat.head()
```

```python
# Membuat list untuk menyimpan data
data_kapal_domestik = []

# Menambahkan data untuk datang
for index, row in kapal_domestik_datang.iterrows():
    data_kapal_domestik.append(['Kapal', 'Datang', row['Provinsi'], 'Domestik', 2022, row['Datang_2022']])
    data_kapal_domestik.append(['Kapal', 'Datang', row['Provinsi'], 'Domestik', 2023, row['Datang_2023']])

# Menambahkan data untuk berangkat
for index, row in kapal_domestik_berangktat.iterrows():
    data_kapal_domestik.append(['Kapal', 'Berangkat', row['Provinsi'], 'Domestik', 2022, row['Berangkat_2022']])
    data_kapal_domestik.append(['Kapal', 'Berangkat', row['Provinsi'], 'Domestik', 2023, row['Berangkat_2023']])

# Membuat DataFrame dari data
tabel_kapal_domestik = pd.DataFrame(data_kapal_domestik, columns=['Kendaraan', 'Status', 'Provinsi', 'Jenis Layanan', 'Tahun', 'Jumlah Penumpang'])
```

```python
tabel_kapal_domestik.head()
```

# **Pesawat Internasional**

# **Integrasi Data**

```python
data_integrasi = pd.concat([tabel_kapal_domestik, tabel_kapal_domestik2, 
                            tabel_pesawat_domestik, tabel_pesawat_domestik2, tabel_pesawat_internasional], ignore_index=True)
data_integrasi
```

```python
data_integrasi.replace("-", "0", inplace=True)
```
